```python
# Node 클래스
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


# LinkedList 클래스
class LinkedList:
    def __init__(self):
        self.head = None
        

    # 맨 뒤에 데이터 삽입
    def append(self, data):
        new_node = Node(data)
        # 리스트가 비어 있는 경우
        if self.head is None:
            self.head = new_node
            return

        # 마지막 노드까지 이동
        current = self.head
        while current.next is not None:
            current = current.next

        # 마지막 노드의 next에 새 노드 연결
        current.next = new_node

    # 맨 앞에 데이터 삽입
    def prepend(self, data):
        new_node = Node(data)

        # 새 노드의 next를 기존 head로 연결
        new_node.next = self.head

        # head를 새 노드로 변경
        self.head = new_node

    # 데이터 삭제
    def delete(self, data):
        # 리스트가 비어 있는 경우
        if self.head is None:
            print("리스트가 비어 있습니다.")
            return

        # 삭제할 데이터가 head에 있는 경우
        if self.head.data == data:
            self.head = self.head.next
            print(f"{data} 삭제 완료")
            return

        # head 이후 노드 탐색
        current = self.head

        while current.next is not None:
            if current.next.data == data:
                current.next = current.next.next
                print(f"{data} 삭제 완료")
                return
            current = current.next

        print(f"{data}를 찾을 수 없습니다.")

    # 데이터 탐색
    def search(self, data):
        current = self.head
        index = 0

        while current is not None:
            if current.data == data:
                print(f"{data}는 {index}번째 위치에 있습니다.")
                return True

            current = current.next
            index += 1

        print(f"{data}를 찾을 수 없습니다.")
        return False

    # 전체 데이터 출력
    def display(self):
        current = self.head

        if current is None:
            print("리스트가 비어 있습니다.")
            return

        while current is not None:
            print(current.data, end=" → ")
            current = current.next

        print("None")
```

```python
# Node 클래스 만들기
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


# 전위 순회: 루트 → 왼쪽 → 오른쪽
def preorder(node):
    if node is None:
        return

    print(node.data, end=" ")
    preorder(node.left)
    preorder(node.right)


# 중위 순회: 왼쪽 → 루트 → 오른쪽
def inorder(node):
    if node is None:
        return

    inorder(node.left)
    print(node.data, end=" ")
    inorder(node.right)


# 후위 순회: 왼쪽 → 오른쪽 → 루트
def postorder(node):
    if node is None:
        return

    postorder(node.left)
    postorder(node.right)
    print(node.data, end=" ")


# 트리 만들기
root = Node("의류")
node_b = Node("상의")
node_c = Node("하의")
node_d = Node("티셔츠")
node_e = Node("블라우스")
node_f = Node("바지")
node_g = Node("스커트")


# 트리 연결하기
root.left = node_b
root.right = node_c

node_b.left = node_d
node_b.right = node_e

node_c.left = node_f
node_c.right = node_g


# 데이터 출력하기
print("전위 순회:")
preorder(root)

print("\n중위 순회:")
inorder(root)

print("\n후위 순회:")
postorder(root)
```

```python
class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    def hash_function(self, key):
        return len(key) % self.size

    def set(self, key, value):
        index = self.hash_function(key)
        self.table[index].append([key, value])

    def get(self, key):
        index = self.hash_function(key)

        for item in self.table[index]:
            if item[0] == key:
                return item[1]
                

        return None

    def show(self):
        for index, data in enumerate(self.table):
            print(index, data)


ht = HashTable(5)

ht.set("apple", "사과")
ht.set("banana", "바나나")
ht.set("melon", "멜론")
ht.set("avocado", "아보카도")
ht.set("tangerine", "귤")
ht.set("Watermelon", "수박")
ht.set("Strawberry", "딸기")
ht.set("orientalmelon", "참외")


ht.show()          

print(ht.get('orientalmelon'))
print(ht.get('Strawberry'))
```

```python
class SimpleHashTable:
    def __init__(self):
        self.size = 5
        self.table = [[] for _ in range(self.size)]

    def hash_function(self, key):
        return len(key) % self.size

    def add(self, key, value):
        index = self.hash_function(key)
        self.table[index].append((key, value))
        print(f"{key}는 index {index}에 저장되었습니다.")

    def show(self):
        for i in range(self.size):
            print(f"index {i}: {self.table[i]}")


hash_table = SimpleHashTable()

hash_table.add("apple", "사과")
hash_table.add("melon", "멜론")
hash_table.add("grape", "포도")
hash_table.add("cat", "고양이")
hash_table.add("dog", "강아지")

hash_table.show()
```

```python
import time

# 0부터 29999까지 숫자가 들어있는 리스트 만들기
num_list = list(range(30000000))

# 29999 중복값 추가
num_list.append(2999)
num_list.append(29999)
num_list.append(299999)
num_list.append(29999999)
# 코드 실행 시작시간 저장
cd_st = time.time()

# 빈 세트 만들기
num_set = set()
# 중복값 확인 여부 - False 없음, True 있음
duplicate = False

# 1~29999까지 있는 리스트 반복문
for i in num_list:
    # 1~29999까지 세트에 있는면 중복여부 확인
    if i in num_set:
        duplicate = True
        print("중복!!!!!!")
        break
    # 중복값이 없으면 num값 추가
    num_set.add(i)
#코드 실행 종료시간 저장
cd_end = time.time()

#코드 실행 여부 및 중복여부 확인
print(f"중복 여부 확인: {duplicate}")
print(f"코드실행 시간 : {cd_end - cd_st} 초")
```

```python
import time
# 0부터 29999까지 숫자가 들어있는 리스트 만들기
numbers = list(range(30000))
# 29999 중복값 추가
numbers.append(29999)
# 코드 실행 시작시간 저장
start_time = time.time()
# 중복값 확인하는 불린값 False는 없음 True 있음
has_duplicate = False

# 중복값이 있는지 확인하는 중복 반복문 1을 29999까지 확인
# 그다음 2를 29999까지 확인 다음 3,4,5,.....29999까지 확인
for i in range(len(numbers)):
    for j in range(i + 1, len(numbers)):
        if numbers[i] == numbers[j]:
            has_duplicate = True

# 코드 실행 종료 시간 저장
end_time = time.time()

# 중복값 찾을때 까지 걸린시간 확인
print("중복 여부:", has_duplicate)
print("실행 시간:", end_time - start_time, "초")
```
