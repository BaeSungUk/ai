```python
import MySQLdb
```

```python
db = MySQLdb.connect(
    host = 'localhost',
    user = 'root',
    password = '1234',
    db = 'fortune_db',
    charset = 'utf8'
)

print(db)
```

```python
cur = db.cursor()

sql = "select * from fortune"

cur.execute(sql)

data = cur.fetchall()

for item in data:
    print(item)
cur.close()
db.close()
```
