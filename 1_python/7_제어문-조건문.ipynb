파이썬의 조건문은 프로그램의 실행 흐름을 제어하기 위한 문법으로, 기본적으로 if, elif, else를 사용하여 조건의 참(True)과 거짓(False)에 따라 다른 코드를 실행할 수 있습니다. 또한 간단한 조건에 대해서는 값1 if 조건 else 값2 형태의 조건부 표현식을 사용해 한 줄로 값을 선택할 수 있으며, Python 3.10부터는 match ~ case 문법을 활용한 구조적 패턴 매칭을 통해 값뿐만 아니라 리스트나 딕셔너리 같은 데이터의 구조를 기준으로 분기 처리가 가능합니다. 이처럼 파이썬은 상황에 따라 다양한 방식의 조건문을 제공하여 간단한 분기부터 복잡한 데이터 처리까지 유연하게 구현할 수 있습니다.

# **1. if, elif, else**

### 1. if 문
가장 기본적인 조건문입니다. 특정 조건이 참(True)일 경우에만 해당 블록 내의 코드가 실행됩니다.
```
if 조건:
    실행할 코드
```    
여기에서 조건은 True 또는 False로 평가될 수 있는 표현식이어야 합니다. 조건이 True로 평가되면 if 문 바로 아래의 코드 블록이 실행됩니다. 그렇지 않으면 그 코드 블록은 건너뜁니다.

````python
age = int(input('나이를 입력하세요.'))

if age % 2 ==0 :
    print('짝수입니다.')
else :
    print('홀수입니다.')
````

````python
user_id = "admin"
password = "1234"
login_fail_count = 0
is_locked = False

while login_fail_count < 3:

    login_id = input("아이디를 입력하세요: ")
    login_password = input("비밀번호를 입력하세요: ")
    
    if is_locked:
        print("계정이 잠겨 있습니다.")

    elif user_id == login_id and password == login_password:
        print("로그인 성공")
        break

    else:
        login_fail_count += 1

        if login_fail_count >= 3:
            print(f"현재 실패 횟수: {login_fail_count}회")
            print("로그인 실패: 계정 잠금")
            is_locked = True # 계정 잠금
        else:
            print("로그인 실패")
            print(f"현재 실패 횟수: {login_fail_count}회")
````

````python
students = {
    "철수": 85,
    "영희": 92,
    "민수": 58,
    "지혜": 74,
    "동현": 40
}

pass_students = set()
fail_students = set()
result_logs = []

# for student in students: 이렇게 for문을 사용하니 딕셔너리에 있는 키값만 출력함.

for name, score in students.items():
    if score >= 80 :
        pass_students.add(name)
        result = '우수'
    elif score >= 60:
        pass_students.add(name)
        result = '통과'
    else : 
        fail_students.add(name)
        result = '불합격'

    result_logs.append((name, score, result))


print(pass_students)
print(fail_students)
print(result_logs)        
````

````python
products = {
    "셔츠": 12,
    "바지": 0,
    "모자": 5,
    "신발": 2,
    "가방": 0
}

sold_out_products = set()
low_stock_products = set()
stock_logs = []

for product, stock in products.items():
    if stock == 0 :
        result = '품절'
        sold_out_products.add(product)
    elif stock <= 5 :
        result = '재고 부족'
        low_stock_products.add(product)
    else :
        result = '재고 충분'
    
    stock_logs.append((product, stock, result))

print(sold_out_products)
print(low_stock_products)
print(stock_logs)    
````
