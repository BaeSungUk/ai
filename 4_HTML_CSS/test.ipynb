```python
def merge_sort(arr):
    # 종료 조건
    # 리스트 길이가 1 이하이면 이미 정렬된 상태
    if len(arr) <= 1:
        return arr

    # 가운데 위치 구하기
    mid = len(arr) // 2

    # 왼쪽, 오른쪽으로 나누기
    left = arr[:mid]
    right = arr[mid:]

    # 왼쪽을 다시 병합 정렬
    left_sorted = merge_sort(left)
    print(f"left_sorted: {left_sorted}")

    # 오른쪽을 다시 병합 정렬
    right_sorted = merge_sort(right)
    print(f"right_sorted: {right_sorted}")

    # 정렬된 왼쪽과 오른쪽을 합치기
    return merge(left_sorted, right_sorted)


def merge(left, right):
    result = []

    i = 0
    j = 0

    # 왼쪽과 오른쪽 둘 다 값이 남아 있을 때 비교
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    # 왼쪽에 남은 값 추가
    while i < len(left):
        result.append(left[i])
        i += 1

    # 오른쪽에 남은 값 추가
    while j < len(right):
        result.append(right[j])
        j += 1

    return result


data = [5, 3, 8, 4, 1 ,22, 14, 17, 30, 24, 2]

print(merge_sort(data))
```
